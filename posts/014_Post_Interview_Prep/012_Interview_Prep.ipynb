## I have a group placement interview for my internship next summer. I am being interviewed by M&A, Consumer and Retail, and TMT groups. Excellent — this is a critical stage in the internship process, and the *group placement* interview is where you show that you’ve gone beyond generic banking prep and have real interest and knowledge about each coverage vertical. Here’s a structured plan with targeted recommendations:

---

## 1. Structure Your “Why This Group?” Answers

Each answer should have **three parts**:

1. **Personal motivation** (why you’re drawn to the industry or type of work),
2. **Skill alignment** (why your strengths fit the group’s workflow), and
3. **Bank-specific angle** (why BofA’s franchise in this group stands out).

### Example: M&A

* **Personal motivation:** “I enjoy working on high-impact, complex situations where strategic reasoning and valuation intersect — M&A gives exposure to board-level decision-making.”
* **Skill alignment:** “I like analytical modeling, synthesis of qualitative strategy and quantitative data, and working under pressure — all central to M&A.”
* **BofA-specific angle:** “BofA has one of the largest M&A platforms globally, advising on cross-border and transformational transactions — such as [insert deal below].”

### Example: Consumer & Retail (C&R)

* **Personal motivation:** “Consumer trends fascinate me — brand loyalty, pricing power, and changing consumer behavior all directly impact valuation.”
* **Skill alignment:** “I enjoy connecting qualitative brand narratives to quantitative performance — which is key in C&R.”
* **BofA-specific angle:** “BofA’s strong relationships with companies like Mondelez, Coca-Cola, and Nike have kept it top-ranked in retail deal flow.”

### Example: TMT

* **Personal motivation:** “I’m drawn to how technology reshapes industries — from AI and semiconductors to streaming.”
* **Skill alignment:** “I like learning fast and adapting — vital in TMT where business models evolve quickly.”
* **BofA-specific angle:** “BofA has been a lead advisor on major telecom and software deals — like the Broadcom–VMware acquisition.”

This shows engagement and initiative.

## 5. Practice the Delivery

* Keep each answer **under 90 seconds** and **story-driven**.
* Be confident but conversational — treat it like a dialogue, not a script.
* Have **one numerical or concrete fact** per answer (e.g., deal size, growth rate, trend stat).
* If you’re asked “rank your preferences,” show clear rationale: “I’m most interested in C&R because of my consumer brand interest, but I’d be excited by M&A or TMT given their analytical and fast-paced environments.”

## I have networking calls with people working in each of those three groups. Provide me with possible questions to ask them related to either the content of their work, industry, or more personal work life questions

Excellent — networking calls are where you demonstrate genuine curiosity and build relationships that can influence your group placement. You want to ask thoughtful questions that go **beyond what’s on Google**, balancing **technical curiosity**, **industry awareness**, and **personal connection**.

Below are **targeted questions** for each group (M&A, Consumer & Retail, and TMT), separated by **three categories**:

1. Content / Nature of Work
2. Industry & Market Trends
3. Personal / Cultural Insight

---

## 💼 M&A (Mergers & Acquisitions)

### 1. Content of Work

* “When you’re working on a live deal, how do analysts typically divide time between modeling, due diligence, and client materials?”
* “What aspects of M&A execution tend to be most challenging or rewarding from your perspective?”
* “How much client interaction do analysts get, especially on smaller or mid-market deals?”
* “Can you walk me through what the internal process looks like after a company decides to explore strategic alternatives?”
* “What kind of modeling or valuation work is most common in your group — LBOs, accretion/dilution, synergies, or all of the above?”

### 2. Industry / Market Trends

* “Has deal flow started to pick up as rates stabilize? Are you seeing more strategic buyers or private equity interest?”
* “What industries have been driving the bulk of your team’s M&A activity recently?”
* “How is AI and data analytics influencing how clients think about acquisition targets or valuation?”
* “Do you see a difference in how clients approach cross-border vs. domestic deals in the current environment?”

### 3. Personal / Work-Life

* “What’s the steepest learning curve you faced when you joined the M&A group?”
* “What makes the culture in BofA’s M&A team unique compared to other banks or groups?”
* “How do you personally manage the intensity of deal timelines?”
* “Was there a moment early on when you realized you were getting real responsibility?”

---

## 🛍️ Consumer & Retail (C&R)

### 1. Content of Work

* “How does your team balance qualitative aspects like brand value and consumer loyalty with quantitative valuation?”
* “How much time do analysts spend tracking company performance versus building materials for live deals?”
* “Do you work more on sell-sides, buy-sides, or strategic advisory engagements?”
* “How do you model brand-driven businesses differently from industrial or tech clients?”
* “Are there differences in how you value global consumer companies vs. domestic ones?”

### 2. Industry / Market Trends

* “What trends have had the biggest impact on consumer M&A over the past year — inflation, supply chains, or digital transformation?”
* “How are clients adapting to shifting consumer preferences, like value-based spending or health/wellness?”
* “Do you see more consolidation in certain subsectors, like food & beverage vs. apparel & luxury?”
* “How do sustainability or ESG factors influence valuation and investor interest in consumer companies?”
* “What differentiates BofA’s C&R franchise when pitching to clients versus competitors like JPM or GS?”

### 3. Personal / Work-Life

* “What do you enjoy most about covering consumer brands versus other industries?”
* “How does your team stay close to consumer trends — are you expected to follow earnings, retail data, or industry reports regularly?”
* “What’s a memorable deal you’ve worked on that captures the essence of C&R banking?”
* “How collaborative is your group across geographies — do you work with London or Asia consumer teams?”
* “What’s the team culture like? Is it more analytical or relationship-driven?”

## 💻 TMT (Technology, Media & Telecom)
