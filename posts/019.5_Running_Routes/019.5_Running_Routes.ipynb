## Prompt 
<span style="background-color: yellow ; padding: 5px;">Please map out running routes, starting from 3700 Walnut Street in Philadelphia for 3, 5, 7, and 10 mile runs that also end at 3700 Walnut street. Make the runs as scenic as possible and give me a few options for each length.</span>

Nice, this is a fun one—running out of 3700 Walnut is elite.

**Assumptions**

* Start/finish = **Pottruck / 3700 Walnut St (37th & Walnut)**.
* You’re fine using bridges + Schuylkill River Trail (SRT) + Fairmount Park.
* Distances are **approximate** (good enough for training, but sanity-check on Strava/MapMyRun).

The Penn Club XCTF site actually has a bunch of standard routes starting/ending at 37th & Walnut, which I’m using as the backbone here. ([Penn Club XCTF][1])

---

## 3-mile options (~3.0–3.3 mi)

### 3A. Campus → Walnut Bridge → SRT out & back (~3.1)

**Scenic:** skyline + river views, almost all on the SRT.

1. Start at **3700 Walnut** and head east on Walnut to **33rd St**.
2. Continue on Walnut over the **Walnut Street Bridge** onto the east bank.
3. Drop down the ramp to the **Schuylkill River Trail** and turn **left (north)**.
4. Run up the trail to just **past the Art Museum area / Paine’s Park skate park**, then turn around.
5. Return south on the SRT, back up the Walnut ramp, and run west on Walnut to 37th.

> Tip: If your watch shows short, add a small campus loop (e.g., up 37th to Chestnut and back).

---

### 3B. Penn Park + South Street Bridge loop (~3.0)

**Scenic:** Penn Park fields, Center City views from South Street Bridge.

1. From 3700 Walnut, go south on **37th** to **Spruce**.
2. Continue east on Spruce to the **entrance of Penn Park**; loop **around the outer path** of Penn Park once.
3. Exit toward the **South Street Bridge**, head up, and cross to the east side.
4. Turn **left** onto the SRT, follow it north to the **Walnut Street ramp**.
5. Run back west on Walnut to 37th.

---

### 3C. Woodlands Cemetery loop (~3.1)

Based on the Penn Club **“Woodlands – 2 miles”** route, just extended a bit on campus. ([Penn Club XCTF][1])

1. From 3700 Walnut, go west on **Walnut** to **40th St**.
2. Turn **left** on 40th, enter **The Woodlands** at the main gate; run a loop on the internal road (do one full perimeter).
3. Exit back on 40th, turn **right** on **Spruce**, then **right** again on **36th**.
4. Take 36th north to Walnut, turn **left** back to 37th.

Add a lap around campus green if you’re short of 3.

---

## 5-mile options (~5.0–5.5 mi)

### 5A. South Street city loop (~5.0)

This is essentially Penn Club’s **“South Street – 5 miles”** route. 

1. From 3700 Walnut, run **east on Walnut** over the Walnut Street Bridge.
2. Continue straight to about **18th–20th St**, then drop south one block to **Chestnut/Market** corridor and work your way to **Broad St** (any east–west street works; use your watch).
3. Head **south on Broad** to **South Street**.
4. Turn **right (west)** on South; follow it all the way to the **South Street Bridge**.
5. Cross the bridge, curve right past HUP/CHOP, then head back to campus via **Spruce → 37th → Walnut**.

Lots of Center City neighborhoods + skyline + river.

---

### 5B. Zoo Loop (~5.0)

Based on the **“Zoo Loop – 5 miles”** map. 

1. From 3700 Walnut, go east on **Market** to **34th**, then down the ramp to **SRT**.
2. Turn **left (north)** and follow the SRT along the river, under the **Art Museum** area and toward the **Zoo / Girard Ave**.
3. At the **Girard Ave / Zoo area**, cross the bridge or use the connecting paths (as on the Zoo Loop map) to switch to the **west bank**.
4. Run **south** along the river path/MLK side until you can cross back (typically near the **Spring Garden** area) or use **Market St** bridge.
5. Return to 37th & Walnut via **Market**.

Mostly river, bridges, and views of Boathouse Row/Museum.

---

### 5C. Art Museum / Rocky Steps loop (~4.5–5.2)

Adapted from the **“Art Museum/Rocky Steps – 4.5 miles”** route; just add a small campus loop. ([Penn Club XCTF][1])

1. From 3700 Walnut, run east via **Walnut or Market** to the SRT.
2. Take the SRT **north** to the **Philadelphia Museum of Art / Rocky Steps**. ([Visit Philadelphia][2])
3. Run up the steps (mandatory) and loop the plaza.
4. Head back along the same SRT path and over Walnut/Market to campus.
5. Add a short loop around **Locust Walk** or the **Quad** area to hit exactly ~5.

---

## 7-mile options (~7.0–7.5 mi)

### 7A. West Fairmount Gardens (~7.0)

This matches Penn Club’s **“West Fairmount Gardens – 7 miles”** route in spirit. 

1. From 3700 Walnut, go east on **Market** to the SRT.
2. Run **north** on the west side of the river, cross onto **MLK/West Fairmount** via a bridge near the Art Museum or upstream.
3. Continue northwest into **West Fairmount Park**, looping around the roads/trails near the **Please Touch Museum** and garden area (as in the West Fairmount Gardens map).
4. Loop back toward the river and return via **the same bridge** and **SRT**.
5. Finish at 37th & Walnut.

Very green, with rolling park roads, fewer cars, and big views over the river.

---

### 7B. MLK via “Black Road” (~7-ish)

A classic Penn workout route (Penn Club calls it **“MLK via Black Road – 7ish miles”**). ([Penn Club XCTF][1])

1. From 3700 Walnut, go to SRT (Walnut/Market).
2. Run **north** on the SRT to the **Art Museum**.
3. Cross to the **MLK Drive** side and follow the river path west/northwest.
4. Turn into **Black Road** (a wooded connector climbing into the park), follow it up to the park roads, then loop back down to MLK.
5. Return the same way via MLK → Art Museum → SRT → Walnut/Market bridge → 37th.

This has a nice mix of flat river path + a solid hill.

---

### 7C. Extended Zoo Loop (~7.0)

Take the 5-mile **Zoo Loop** and push it a bit farther upriver.

1. Follow **Zoo Loop** directions to the **Zoo/Girard** area. 
2. Instead of crossing back immediately, continue further **up Kelly Drive or MLK Drive** (pick one side) until your watch hits ~3.5 miles.
3. Turn around and come back the same way, crossing back to Penn via Walnut/Market as before.

The extra section north of the Zoo is pretty, quiet, and very “Philly runner” core.

---

## 10-mile options (~10–10.5 mi)

### 10A. Strawberry Mansion Bridge loop (~10.5)

Straight from Penn Club’s **“Strawberry Mansion – 10.5 miles”** route. 

1. From 3700 Walnut, take **Market → SRT** and head **north** along the river.
2. Pass the **Art Museum**, Boathouse Row, and continue up **Kelly Drive** on the river path. ([DiscoverPHL][3])
3. Continue all the way to the **Strawberry Mansion Bridge**, cross to the opposite bank.
4. Return downstream on the **MLK / East Fairmount Park side**, eventually crossing back at one of the bridges (often near the Art Museum).
5. Rejoin the SRT and run back to Walnut/Market → 37th.

Water, trees, boathouses, bridges—super scenic and iconic.

---

### 10B. Kelly Drive East Falls loop (~9–10 mi from Penn)

The classic **Kelly Drive “East Falls” loop** is ~8.4–9 mi from the Art Museum; from Penn it’s just over 9–10. ([DiscoverPHL][3])

1. Run from 3700 Walnut to the **Art Museum** via SRT (~2 mi).
2. Continue up **Kelly Drive** path along the river to **Falls Bridge**.
3. Cross Falls Bridge, turn **left**, and come back on the **MLK side** of the river to the Art Museum.
4. Finish by returning to Penn via SRT and Walnut/Market.
5. If you’re short of 10, add a small loop around **Penn Park** before finishing.

---

### 10C. Fairmount Park figure-eight (~10 mi)

Combine West Fairmount + Strawberry Mansion vibes.

1. From 3700 Walnut → SRT → north to **Art Museum**.
2. Cross to the **MLK side**, head up to near **Black Road**, climb into **West Fairmount Park**, and loop around the **Please Touch Museum / Belmont Plateau** roads. 
3. Drop back to the river and continue north to **Strawberry Mansion Bridge**, cross, then head back down **Kelly Drive**.
4. Finish via SRT back to Walnut/Market → 37th.

Use your watch to trim loops in West Fairmount so you land around 10.

## Final notes

* **Safety:**

  * Use lights/reflective gear at night, watch traffic at bridge ramps, and be careful on some darker park segments.
* **Measurement:**

  * These are **approximate**, but all are standard runner loops; plug into **Strava / MapMyRun / Garmin** starting from “Pottruck / 3700 Walnut” to get exact mileage.
* **Weather & river:**

  * After storms, some SRT sections can be slick or briefly flooded—worth checking local reports. ([Schuylkill Banks][4])
