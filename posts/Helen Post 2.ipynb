### **Final Tips for Student Dining & Budgeting**

* Try eating lunch instead of dinner when possible — many places have better value midday.
* Share dishes with a friend: especially at the $20+ places, splitting mains or desserts helps.
* Use off-peak times (before 6 pm) to avoid crowds and get better seating.
* Ask about student discounts or early-bird specials; some places offer them.
* Explore the many food-carts around campus for ultra-budget meals (<$10).
* Keep variety in mind: mix cuisines, price-points, and settings so you don’t get food fatigue.
